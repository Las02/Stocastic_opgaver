```python
import numpy as np
import pandas as pd
import scipy
from icecream import ic
from matplotlib import pyplot as plt
import seaborn as sns
from tqdm import tqdm
```

# 1.a Exponential Distribution

```python

# Parameters
scale = 1.0  # scale parameter (1/Î»)
size = 10000  # number of samples

# Generate exponential distribution
exp_samples = np.random.exponential(scale, size)

# Plot
plt.hist(exp_samples, bins=50, density=True, alpha=0.6, color='g')
plt.title('Exponential Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

# 1.b Normal Distribution

```python
# Parameters
size = 10000  # number of samples

# Generate uniform random numbers
u1 = np.random.rand(size)
u2 = np.random.rand(size)

# Box-Muller transform
z0 = np.sqrt(-2 * np.log(u1)) * np.cos(2 * np.pi * u2)
z1 = np.sqrt(-2 * np.log(u1)) * np.sin(2 * np.pi * u2)

# Plot
plt.hist(z0, bins=50, density=True, alpha=0.6, color='b')
plt.title('Normal Distribution (Box-Muller)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

# 1.c Pareto Distribution

```python
# Parameters
k_values = [2.05, 2.5, 3, 4]  # different values of k
size = 10000  # number of samples
scale = 1  # scale parameter, Î² = 1

for k in k_values:
    # Generate Pareto distribution
    pareto_samples = (np.random.pareto(k, size) + 1) * scale

    # Plot
    plt.hist(pareto_samples, bins=1000, density=True, alpha=0.6, label=f'k={k}')

plt.title('Pareto Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.xlim(0, 10)
plt.legend()
plt.show()
```
